```python
import pandas as pd
```

```python
listings=pd.read_csv('listings_eat2.csv')
```

```python
listings.info()
```

```python
listings.head()
```

```python
columns=['Original Price','Sale Price']
for col in columns:
    listings[col]=listings[col].str.replace('S','')
    listings[col]=listings[col].str.replace('$','')
    listings[col]=pd.to_numeric(listings[col])
listings.info()
```

```python
listings.drop(columns='Unnamed: 0')
```

```python
listcount=pd.DataFrame(listings['Company Name'].value_counts())
listcount.reset_index(inplace=True)
```

```python
listcount.columns=['name','count']
listcount.info()
def getcount(name):
    for i in range(len(listcount)):
        if name==listcount['name'][i]:
            return listcount['count'][i]
```

```python
len(listcount['name'])
```

```python
listings['listing count']=listings['Company Name'].apply(lambda x:getcount(x))
```

```python
print(listings['Num_bought'].max(), listings['Num_bought'].min())
```

```python
def popularity(numbought):
    if numbought<100:
        return "Below 100"
    elif numbought <200:
        return "100 - 199"
    elif numbought <300:
        return "200- 299"
    elif numbought <400:
        return "300- 399"
    elif numbought <500:
        return "300- 499"
    elif numbought <1000:
        return "500- 999"
    elif numbought <2000:
        return "1000- 1999"
    else:
        return "2000 and above"
```

```python
listings['popularity']=listings['Num_bought'].apply(popularity)
```

```python
listings.groupby('popularity').mean()
```
